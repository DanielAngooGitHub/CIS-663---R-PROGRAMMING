```python
import pandas as pd

# Initialize an empty DataFrame to store customer data
columns = ["Name", "Email", "Phone Number", "Address", "Date of Birth", "Account Type"]
customer_data = pd.DataFrame(columns=columns)

def collect_customer_data():
    """Collect prospective customer data for account opening."""
    print("Welcome to the Bank's Customer Data Collection System!")

    while True:
        # Collect customer information
        name = input("Enter customer's full name: ").strip()
        email = input("Enter customer's email address: ").strip()
        phone = input("Enter customer's phone number: ").strip()
        address = input("Enter customer's address: ").strip()
        dob = input("Enter customer's date of birth (YYYY-MM-DD): ").strip()
        account_type = input("Enter account type (Savings/Current): ").strip().capitalize()

        # Validate inputs (basic checks)
        if not name or not email or not phone or not address or not dob or account_type not in ["Savings", "Current"]:
            print("\nInvalid input. Please fill all fields correctly.\n")
            continue

        # Append data to DataFrame
        global customer_data
        customer_data = pd.concat([
            customer_data,
            pd.DataFrame([[name, email, phone, address, dob, account_type]], columns=columns)
        ], ignore_index=True)

        print("\nCustomer data collected successfully!")

        # Ask if user wants to add another customer
        another = input("Do you want to add another customer? (yes/no): ").strip().lower()
        if another != 'yes':
            break

def display_data():
    """Display the collected data."""
    print("\nCollected Customer Data:")
    print(customer_data)

def save_data_to_file():
    """Save the collected data to a CSV file."""
    filename = "customer_data.csv"
    customer_data.to_csv(filename, index=False)
    print(f"\nCustomer data saved to {filename}")

def main():
    """Main application flow."""
    while True:
        print("\n--- Bank Customer Data Collection System ---")
        print("1. Collect Customer Data")
        print("2. View Collected Data")
        print("3. Save Data to File")
        print("4. Exit")

        choice = input("Enter your choice: ").strip()

        if choice == '1':
            collect_customer_data()
        elif choice == '2':
            display_data()
        elif choice == '3':
            save_data_to_file()
        elif choice == '4':
            print("\nThank you for using the Bank's Customer Data Collection System!")
            break
        else:
            print("\nInvalid choice. Please select a valid option.\n")

if __name__ == "__main__":
    main()
```
